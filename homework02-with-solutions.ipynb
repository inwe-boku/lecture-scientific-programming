Using matplotlib, plot a histogram of the respective generation values, i.e. the output of `windturbine_simulation_improved_mw`. Use google to find the correct function to plot a histogram.

Let's assume that this histogram is a good approximation of the wind power generation at our turbine locations.

Using google, determine how to calculate the mean of a numpy array. Calculate the mean and store it in `wind_power_mean_mw`. Furthermore, calculate the total annual production of the turbine in MWh per year based on the mean production by multiplying the mean with the number of hours in a year and store the resulting value in `wind_energy_total_mwh_per_yr`?

```python3
import numpy as np
import matplotlib.pyplot as plt


def windturbine_simulation_improved_mw(wind_speed_ms, rotor_diameter_m, rated_capacity_mw):
    """Calculate output of a wind turbine in MW, given wind speed in m/s and rotor_diameter in m, 
       accounting for rated capacity."""
    c_p = 0.4
    rho = 1
    area = rotor_diameter_m**2 * np.pi / 4
    p_out = c_p * 0.5 * rho * area * wind_speed_ms**3 / 1e6
    p_out = np.clip(p_out, 0, rated_capacity_mw)
    return p_out
```

```python3
wind_speeds_ms_stuhleck = np.array([3, 2, 4, 11, 7, 13, 12, 9, 3, 7, 11, 2, 7, 4, 7, 4, 8, 5, 9, 9])

rotor_diameter_vestas = 100
rated_capacity_vestas = 3
electrolyzer_efficiency = 0.65

wind_powers_stuhleck = windturbine_simulation_improved_mw(wind_speeds_ms_stuhleck, rotor_diameter_vestas, rated_capacity_vestas)

plt.hist(wind_powers_stuhleck)
plt.xlabel('Output power (MW)')
plt.ylabel('Frequency')

wind_power_mean_mw = np.mean(wind_powers_stuhleck)
wind_energy_total_mwh_per_yr = np.mean(wind_powers_stuhleck) * 8760
```

```python3
def max_of_two(x, y):
    if x > y:
        return x
    else:
        return y

def max_of_three(x, y, z):
    return max_of_two(x, max_of_two(y, z))
```
