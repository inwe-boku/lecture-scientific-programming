# Recap functions

Functions are blocks which encapsulate code. The code is executed when the function is called. Functions may have no, one or several parameters. These represent the input to the function. A function may return a value upon completion.

Functions can be written by you. But there are of course also a lot of pre-existing functions in Python - and by using modules, you can have access to even more functions.

```python
import numpy as np

# function definition
def load_minus_renewables_mw(load_mw, renewable_generation_mw):
    return load_mw - renewable_generation_mw

load_tx_mw = np.array([70, 75, 82, 79, 65])
renewable_generation_tx_mw = np.array([14, 17, 21, 15, 13])

# function call
print(load_minus_renewables_mw(load_tx_mw, renewable_generation_tx_mw))

# double renewable capacity
renewable_generation_tx_double_mw = 2 * renewable_generation_tx_mw

# another function call
print(load_minus_renewables_mw(load_tx_mw, renewable_generation_tx_double_mw))
```

## Exercise 5

Write a function `center_value(l)` which takes a numpy array as input and returns the value in the middle, e.g. if an array `np.array([1, 42, 3])` is passed to the function it should return `42`.

Are there any ambiguities or other problems with this specification? Which reasonable options are there to deal with corner cases? For which cases is there an obvious solution and for which cases not?

```python
### BEGIN SOLUTION
```

```python
def center_value(a):
    # for empty lists, we raise an error - probably more reasonable than return None or NaN
    if len(a) == 0:
        raise ValueError("a is of length 0, cannot find a center element in an empty list")

    # if the list contains an even nukmber of elements, let's just pick the one with lower index
    # we could of course also pick the one with higher index, but doesn't make a difference
    # another reasonable option: calculate the mean of the two center values, but who said that the elements must be numbers?
    # could be a list of strings or lists!
    center_index = int((len(a) - 1) / 2)
    return a[center_index]
```

```python
center_value(np.array([1, 42, 3]))
```

```python
center_value(np.array([1, 42]))
```

```python
center_value(np.array([]))
```
